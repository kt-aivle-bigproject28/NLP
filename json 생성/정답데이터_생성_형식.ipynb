```python
import os
def generate_json_from_txt(txt_name, input_dir, output_dir):
    text_data_path = os.path.join(input_dir, f"{txt_name}.txt")
    json_output_path = os.path.join(output_dir, f"{txt_name}.json")
    
    # JSON 구조 정의
    json_structure = created_json
    
    # 텍스트 파일 읽기
    with open(text_data_path, 'r', encoding='utf-8') as file:
        json_structure["text_data"] = file.read()
    
    # 디렉토리 생성 (필요한 경우)
    os.makedirs(output_dir, exist_ok=True)
    
    # JSON 파일 쓰기
    with open(json_output_path, 'w', encoding='utf-8') as json_file:
        json.dump(json_structure, json_file, ensure_ascii=False, indent=4)
    
    print(f"JSON 파일 생성 완료: {json_output_path}")

# 실행
txt_name = "화폐전쟁 3 (쑹훙빙) (Z-Library)"
input_dir = r"C:\Users\User\Desktop\데이터_2만자\0108_데이터(앞부분 날림 + 2만자)"
output_dir = r"C:\Users\User\Downloads"

generate_json_from_txt(txt_name, input_dir, output_dir)
```
